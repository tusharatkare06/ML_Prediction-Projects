```python
from catboost import CatBoostClassifier, Pool
import catboost as cb
```

```python
def get_categorical_indicies(X):
    cats = []
    for col in X.columns:
        if pd.api.types.is_numeric_dtype(X[col]):
            pass
        else:
            cats.append(col)
    cat_indicies = []
    for col in cats:
        cat_indicies.append(X.columns.get_loc(col))
    return cat_indicies
categorical_indicies = get_categorical_indicies(X2)
```

```python
categorical_indicies
```

```python
train_dataset = cb.Pool(X2_train,y2_train, 
                        cat_features=categorical_indicies)                                                      
test_dataset = cb.Pool(X2_test,y2_test,           
                       cat_features=categorical_indicies)
```

```python
def convert_cats(X):
    cats = []
    for col in X.columns:
        if pd.api.types.is_numeric_dtype(X[col]):
            pass
        else:
            cats.append(col)
    cat_indicies = []
    for col in cats:
        X[col] = X[col].astype('category')
convert_cats(X2)
```

```python
model_cb = cb.CatBoostClassifier(loss_function='Logloss',  
                              eval_metric='Accuracy')
```

```python
model_cb.fit(train_dataset)
```
